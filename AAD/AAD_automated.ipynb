```python
import pandas as pd
```

```python
# --- Load the two Excel files ---
file1 = "data/17636838404621cba7a5.xlsx"
file2 = "data/17636838682589182320.xlsx"
```

```python
df1 = pd.read_excel(file1)
df2 = pd.read_excel(file2)

# --- Assuming the column containing names is called "Name" ---
col = "Name"

# Normalize for comparison (lowercase)
set1 = set(df1[col].str.lower().str.strip())
set2 = set(df2[col].str.lower().str.strip())

# Names in both files
common = set1.intersection(set2)

# Names only in file1 or file2
unique_to_file1 = set1 - set2
unique_to_file2 = set2 - set1

# Print results
print("Names appearing in BOTH files:")
for name in common:
    print(name)

print("\nNames ONLY in File1:")
for name in unique_to_file1:
    print(name)

print("\nNames ONLY in File2:")
for name in unique_to_file2:
    print(name)
```
