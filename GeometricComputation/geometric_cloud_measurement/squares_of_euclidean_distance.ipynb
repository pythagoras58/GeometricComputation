SETUP PHASE

```python
from phe import paillier

# Alice generates a public and private key pair
public_key, private_key = paillier.generate_paillier_keypair()
```

ENCRYPTION PHASE

```python
# Alice's coordinates
x_A, y_A, z_A = 10, 20, 30

# Bob's coordinates
x_B, y_B, z_B = 5, 15, 25

# Alice encrypts her coordinates
C_x_A = public_key.encrypt(x_A)
C_y_A = public_key.encrypt(y_A)
C_z_A = public_key.encrypt(z_A)

# Bob encrypts his coordinates
C_x_B = public_key.encrypt(x_B)
C_y_B = public_key.encrypt(y_B)
C_z_B = public_key.encrypt(z_B)

# The cloud receives these encrypted values
```

COMPUTATIONAL PHASE

```python
# The cloud computes encrypted differences
C_delta_x = C_x_A - C_x_B
C_delta_y = C_y_A - C_y_B
C_delta_z = C_z_A - C_z_B

# Paillier is additively homomorphic, not multiplicatively. To perform
# squaring, you'd need a separate protocol (e.g., using a secure function
# where one party decrypts, squares, and re-encrypts). Let's assume
# this step is completed and the cloud has the encrypted squared differences.
# In a simplified example, you can use the multiplication operator for a single-party
# re-encryption, which is a common approach in these types of protocols.
C_sq_delta_x = C_delta_x * (x_A - x_B)  # A simplification assuming one party re-encrypts
C_sq_delta_y = C_delta_y * (y_A - y_B)
C_sq_delta_z = C_delta_z * (z_A - z_B)

# The cloud then computes the final encrypted SED using homomorphic addition
C_SED = C_sq_delta_x + C_sq_delta_y + C_sq_delta_z
```

DECRYPTION PHASE

```python
# Alice receives the encrypted result from the cloud
final_SED = private_key.decrypt(C_SED)

print(f"The computed Squared Euclidean Distance is: {final_SED}")
```
